```
#Upload image function/method
import requests
import json
import base64
import urllib.request
import os

#There will be global variables for auth data
WPBaseURL = 'https://whatbookshouldiread.com' # the url of the wp access loc
user = 'jordansnider' #API Service User
appPass = 'yvIX tvBJ 8pF6 lrZW Anib mv2T' #App password for serviceUser
token = base64.standard_b64encode((user + ':' + appPass).encode('utf-8')) # we have to encode the usr and pw

def upload_image_wordpress(imageURL,isbn):
  """Upload an image to WordPress, accepting an image URL and ISBN#"""
  #First Wwe download the image provied by URL
  response = urllib.request.urlretrieve(imageURL, "{0}.jpg".format(isbn)); #get, save image locally
  imagePath = response[0]; #save the path of the locally saved image
  if imagePath == "":
    print("Failed to download image by URL");
    return None
  print('image saved locally via: ', imagePath);  

  #Next we then take locally downloaded image and upload it to WordPress
  headers = {'Authorization': 'Basic ' + token.decode('utf-8')} # here we define which the auth. type used
  media = {'file': open(imagePath,'rb')} #Load as a binary file
  image = requests.post(WPBaseURL + '/wp-json/wp/v2/media/', headers=headers, files=media) #create a post request with the file
  if image == "":
    print("Failed to upload image to WordPress")
    return None
  link = json.loads(image.content.decode('utf-8'))['link'] # get the link to the image out of the response
  postid =json.loads(image.content.decode('utf-8'))['id'] # get the post-id of the image out of the response (we will use it further down to change the wp-image parameters)
  print('Your image is published on {0} with ID {1}'.format(link, postid))

  #PUT update the image with relevant meta info
  post = {'caption': '(BookTitleHere) book cover',
        'description': '(BookTitleHere) book cover'
        }
  r = requests.post(WPBaseURL + '/wp-json/wp/v2/media/'+str(postid), headers=headers, json=post) # make now a standard post request this time with the json 'post' as content
  if r == "":
    print("Failed to update image metadata");
    return None
  print('Your image is updated on {} with ID {}'.format(link, postid)) 

  #delete the image from local storage once we are done with it
  os.remove(imagePath);
  print("image safely removed from local os:", imagePath);
  return True;


#Variables to fill the gap of an actual book object
isbn = "9780333376966";
bookImageURL = "https://images.isbndb.com/covers/69/66/9780333376966.jpg";
bookUpload = upload_image_wordpress(bookImageURL,isbn);
print(bookUpload);
```
